# Phase 2 & 3: Data Discovery and Engineering

In this notebook, we mock the ingestion of data from sources like the World Bank, UN, and IEA, perform data cleaning, treat missing values, and engineer features.

```python
import pandas as pd
import numpy as np
import os

# Mock Data Loading
print("Loading datasets from World Bank, UN, and IEA...")
np.random.seed(42)
years = np.arange(2000, 2024)
countries = ['USA', 'CHN', 'IND', 'NGA', 'CMR', 'DEU']

# Create dummy dataframe
data = []
for c in countries:
    for y in years:
        data.append({
            'country': c,
            'year': y,
            'gdp': np.random.uniform(1e10, 2e13),
            'co2_emissions': np.random.uniform(10, 10000),
            'population': np.random.uniform(1e7, 1.4e9)
        })
df = pd.DataFrame(data)
df.head()
```

## Missing Value Treatment & Outlier Detection

```python
# Impute missing values using forward fill
df.fillna(method='ffill', inplace=True)

# Engineer Feature: CO2 per capita
df['co2_per_capita'] = df['co2_emissions'] / df['population']
print("Data Engineering Complete. Dataset shape:", df.shape)
```
